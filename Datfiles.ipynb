```python
import numpy as np
import pandas as pd
import csv
from matplotlib import pyplot as plt
```

```python
data = np.genfromtxt('psl240nm_220619193306.dat',
                     skip_header=15,
                     skip_footer=1,
                    
                     dtype=None,
                     delimiter='\t', encoding='ascii')
```

```python
data.dtype.names
```

```python
Voltage = data['f2']
MobilityDia = data['f4']
Numc = data['f5']
```

```python
import matplotlib.pyplot as plt


```

```python
fig, axs = plt.subplots(2, ncols=1,figsize=(10,8))
axs[0].plot(Voltage,Numc , ':*')
axs[1].plot(MobilityDia, Numc, ':*')
axs[0].set(xlabel="DMA Voltage (Volts) ",ylabel="Num. Conc. #/cc")
axs[0].grid()
axs[1].set(xlabel="Mobility Diameter nm",ylabel="Num. Conc. #/cc")
axs[1].grid()
axs[0].text(0.1, 0.9, 'Seath Flow = 3.7LPM', horizontalalignment='center', verticalalignment='center', transform=axs[0].transAxes)
plt.savefig('DMA4')
```

```python
# Import required packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pylab import cm
font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }
```

```python
SN, CV, Dmavolt, mob, mob_dia, numc, t = np.loadtxt('psl240nm_220619193306.dat', unpack=True, delimiter='\t', skiprows=15)
```

```python
# Create figure and add axes object
fig = plt.figure()



ax1 = fig.add_axes([0, 0, 0.6, 0.6])
ax1.plot(Dmavolt, numc, ':*')

ax1.xaxis.set_tick_params(which='major', size=5, width=2, direction='in', top='off')
#ax1.xaxis.set_tick_params(which='minor', size=5, width=2, direction='in', top='off')
ax1.yaxis.set_tick_params(which='major', size=10, width=2, direction='in', right='off')
#ax1.yaxis.set_tick_params(which='minor', size=7, width=2, direction='in', right='on')
ax1.grid()
#ax1.set_xlim(0, 4000)
ax1.set_xlabel('DMA Voltage',fontsize= 14, fontdict = font ) 
#ax1.set_ylim(0, 18000)
ax1.set_ylabel('Concentration [#/cc]', fontsize =14, fontdict = font)




ax2 = fig.add_axes([0, 0.8 , 0.6, 0.6]) #origin, height, width of figure
ax2.plot(mob_dia, numc, ':*')
ax2.xaxis.set_tick_params(which='major', size = 5, width=2, direction='in', top = 'off')
ax2.yaxis.set_tick_params(which='major', size= 5, width = 2, direction ='in', top ='off')
ax2.grid()
#ax2.set_xlim(0,500)
#ax2.set_ylim(0, 18000)
ax2.set_xlabel('Mobility Diameter [nm]', fontsize = 14, fontdict = font)
ax2.set_ylabel('Concentration [#/cc]', fontsize = 14, fontdict = font)
# Save figure

plt.savefig('Final_Plot.png', 
            dpi=1000, 
            bbox_inches='tight', 
            pad_inches =1, 
            transparent = True, 
            facecolor = 'w', 
            edgecolor = 'w',
            orientation = 'landscape')
plt.show()
```
